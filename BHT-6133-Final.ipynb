```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
rng = np.random.default_rng()
x = rng.integers(101, size=(100, 100))
x
```

```python
sns.heatmap(x, cmap = 'hot')
plt.title("Random Integers Heatmap")
plt.savefig("heatmap1.png", dpi=600)
plt.show()
```

```python
odd_numbers = x % 2 == 1
odd_numbers
```

```python
x[odd_numbers]
```

```python
y = np.multiply(odd_numbers, 1)
y
```

```python
sns.heatmap(y, cmap = 'hot')
plt.title("True/False Heatmap")
plt.savefig("heatmap2.png", dpi=600)
plt.show()
```
